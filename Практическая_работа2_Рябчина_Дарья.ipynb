```python
# Метод для вывода показаний barchart. 
# Источник: https://www.statology.org/seaborn-barplot-show-values/

def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

#### Тестирование гипотез. Непараметрические критерии

Для начала сформулируем гипотезы для проверки:\
H0: распределения по переменной Gender и satisfaction не зависимо.\
H1: распределения по переменной Gender и satisfaction зависимо.

```python
import scipy.stats

# импортируем библиотеку для расчета криетрия chi2
```

```python
alpha = 0.05

# выберем уровень значимости
```

```python
ct_o = hotels.groupby('Gender')['satisfaction'].value_counts().unstack()
ct_o

# построим таблицу сопреженности
```

```python
dof = (2-1)*(2-1)
dof

# рассчитаем степени свободы
```

```python
critical_value = scipy.stats.chi2.ppf(0.95,1)
critical_value

# рассчитаем наше критические значение для критерия chi2 
```

```python
chi_2 = scipy.stats.chi2_contingency(ct_o, correction = False)
chi_2

# рассчитаем хи2
```

```python
if chi_2[0] >= critical_value:
    print ('Отвергаем HO')
else:
    print ('Не отвераегм H0')
    
# зададим условием для того, чтобы подтвердить или опровергунть нашу гипотезу
```

Таким образом, мы можем сделать вывод о том, что существует зависимость между переменными Gender и satisfaction.

```python
v_1 = var_1/n_1
v_2 = var_2/n_2
dof_welch = (v_1 + v_2)**2 / (v_1**2/(n_1 - 1) + v_2**2/(n_2 - 1))
print(int(dof_welch))

# рассчитываем степень свободы с поправкой welch 
```

```python
t_critical = scipy.stats.t.ppf(0.95, 96155)
print(t_critical)

# рассчитаем наше критические значение для критерия t 
```

```python
t = scipy.stats.ttest_ind(women, men, equal_var = False)
print (t)

# рассчитаем наш критерий t
```

```python
if abs(t[0]) > t_critical:
    print ('Отвергаем HO')
else:
    print ('Не отвераегм H0')
    
# зададим условием для того, чтобы подтвердить или опровергунть нашу гипотезу
```

Таким образом, мы делаем вывод о том, что существует разница между тем, как выставляют оценки отелю мужчины и женщины (из чего складывается итоговая оценка). Также стоит отметить, что у нас достаточно маленькое значение pvalue, а это значит, что у нас достаточно низкие шансы получить ошибку 1-го типа.

Рассчитаем доверительный интервал

```python
s_m1_m2 = np.sqrt(v_1 + v_2)
print(s_m1_m2)

# рассчитаем стандартную ошибку
```

```python
mean_1 = women.mean()
mean_2 = men.mean()

# рассчитаем средние наших выборок
```

```python
CI_bottom = (mean_1 -  mean_2) - t_critical*s_m1_m2
CI_top = (mean_1 -  mean_2) + t_critical*s_m1_m2

print(CI_bottom)
print(CI_top)

# рассчитаем доверительный интервал
```

Мы можем с 95% вероятностью утверждать, что разница между средними по нашим выборкам находится в интервально между -0.23 и -0.08. Отрицательные значения связаны с тем, что среднее по первой выборке меньше, чем средннее по второй выборке.

#### Линейная регресия

```python
sat_mean = hotels.groupby(hotels['Age'])['satisfaction'].mean()
loc_mean = hotels.groupby(hotels['Age'])['Hotel location'].mean()
dep_arr_mean = hotels.groupby(hotels['Age'])['Departure/Arrival  convenience'].mean()
booking_mean = hotels.groupby(hotels['Age'])['Ease of Online booking'].mean()
food_mean = hotels.groupby(hotels['Age'])['Food and drink'].mean()
check_mean = hotels.groupby(hotels['Age'])['Checkin/Checkout service'].mean()
other_mean = hotels.groupby(hotels['Age'])['Other service'].mean()
wifi_mean = hotels.groupby(hotels['Age'])['Hotel wifi service'].mean()
clean_mean = hotels.groupby(hotels['Age'])['Cleanliness'].mean()
comf_mean = hotels.groupby(hotels['Age'])['Stay comfort'].mean()
ent_mean = hotels.groupby(hotels['Age'])['Common Room entertainment'].mean()

hotels_ages = pd.concat([sat_mean, loc_mean, dep_arr_mean, booking_mean, food_mean, check_mean, other_mean, wifi_mean, clean_mean, comf_mean, ent_mean], axis = 1)
hotels_ages.head()

# для построения линейной регресии агрегируем наши данные по возрасту респондентов
```

```python
plt.figure(figsize=(20,20))
p=sns.heatmap(hotels_ages.corr(), annot=True, cmap='PuOr', vmin=-1, vmax=1)

# проверим мультиколлинеарность наших данных
```

Таким образом, наболее сильной корреляцией в датасете является корреляция между Common Room entertainment и satisfaction.

```python
sns.pairplot(hotels_ages)

# проверим наличие линейной взаимосвязи
```

На основе графиков рассеивания мы также можем увидеть, что между Common Room entertainment и satisfaction есть линейная зависимость и мы можем построить линейную регрессию для этих переменных.

```python
X = hotels_ages['Common Room entertainment']
y = hotels_ages['satisfaction']

res = scipy.stats.linregress(X,y)
print(res)

# построим модель линейной регрессии
```

Итак, мы получили угол наклона (slope), а точнее значение нашего независимого предиктора равное 0,692: при одно шаге в Common Room entertainment значение satisfaction увеличивается на 0,692. Intercept -  значение для предиктора рановго 0, при котором линия регрессии пересекает ось y. Также значение pvalue  говорит о том, что наша связь статистически значима.

```python
r_sq= res.rvalue ** 2
r_sq

# вычислим коэффициент детерминации
```

То есть можно сказать, что наша модель достаточно хорошая, так как мы получили достаточно высокий коэффициент детерминации. Наша модель на 84% объясняет дисперсию в целевой переменной.

```python
plt.scatter(X, y, label = 'оригинальные данные')
plt.plot(X, res.intercept + res.slope * X, 'g', label = 'линия регрессии')
plt.legend()
plt.show()

# построим график, где отобразим линию регрессии на графике рассеивания для выбранных переменных
```

```python
import statsmodels.api as sm

# импортируем библиотку для построения модели линейной регрссии с несколькими предикторами
```

```python
X_simple = hotels_ages['Common Room entertainment']
y = hotels_ages['satisfaction']
X_simple_constant = sm.add_constant(X_simple)

model = sm.OLS(y, X_simple_constant)
res_simple = model.fit()
print(res_simple.summary())

# строим модель линейной регрессии с помощью statsmodels
```

По результатам модели мы видим, что оба наших коэффицента статистически значимы, а значения при свободном члене и при X равны значениям, рассичитанным в предущей модели.
Коэффициент детерминации также демонстрирует достаточно высокое значение и очень близок по значению в предыдщуей модели. 

```python
X_mult = sm.add_constant(hotels_ages.drop('satisfaction', axis=1))

model = sm.OLS(y, X_mult)
res_mult = model.fit()
print(res_mult.summary())

# строим модель линейной регрессии с несколькими предикторами с помощью statsmodels
```

В данной модели с несколькими предикторами коэффициент детерминации увеличился, что говорит нам об улучшении нашей модели благодаря введению новых переменных. Также некоторые коэффициенты не статистически значимы: Hotel location, Food and drink, Checkin/Checkout service, Common Room entertainment. При этом некоторые переменные идут с отрицательным коэффициентом: так, получается, что чем хуже сервис wifi в отеле, тем выше коэффициент удовлетворенности услугами отеля.
